
DSA Practice Activities




Array



Easy


1. Reverse an Array


Write a program that asks the user for the array size and its elements, then creates a function to reverse the array.

•	Parameters: an integer array, and the size of the integer array

•	Return type: an integer array (reversed array)

•	Return value: the reversed array

•	Inputs: array size and the succeeding elements

•	Expected output: reversed array elements


2. Sum of Array Elements


Write a program that asks the user for the array size and its elements, then creates a function to calculate the sum of the elements.

•	Parameters: an integer array, and the size of the integer array

•	Return type: int

•	Return value: the sum of the array elements

•	Inputs: array size and the succeeding elements

•	Expected output: sum of the array elements

Medium


3. Matrix Addition


Write a program that asks the user for the size and elements of two matrices, then creates a function to add the two matrices.

•	Parameters: two 2D integer arrays, the number of rows, and the number of columns

•	Return type: a 2D integer array (resultant matrix)

•	Return value: the resultant matrix after addition

•	Inputs: number of rows, number of columns, and the elements of both matrices

•	Expected output: resultant matrix elements


4. Binary Search


Write a program that asks the user for the array size and its elements, then creates a function to perform a binary search on the array.

Parameters: an integer array, the size of the integer array, and the target element

•	Return type: int

•	Return value: the index of the target element if found, otherwise -1

•	Inputs: array size, the succeeding elements, and the target element

•	Expected output: index of the target element if found, otherwise -1

Hard


5. Linked List Implementation


Write a program to create a singly linked list and perform various operations on it such as insertion, deletion, and traversal.

•	Operations: Insert at beginning, insert at end, delete a node, traverse the list

•	Expected output: List elements after each operation


6. Depth-First Search (DFS) and Breadth-First Search (BFS) on Graph


Write a program to create a graph and perform DFS and BFS on it.

•	Parameters: an adjacency list of the graph and the starting vertex

•	Return type: a list of vertices visited in the order of traversal

•	Return value: list of vertices visited in DFS and BFS

•	Inputs: number of vertices, number of edges, and the edges of the graph

•	Expected output: list of vertices visited in DFS and BFS order




Dynamic Data Structure



Easy



1. Dynamic Array Allocation


•	Task: Write a program that dynamically allocates an array of integers. Ask the user for the array size and its elements, and then print the elements.

•	Inputs: Array size and the succeeding elements.

•	Expected Output: The elements of the array.


2. Struct for Student Information


•	Task: Create a struct called Student with fields for name, age, and GPA. Dynamically allocate an array of Student structs. Ask the user for the number 
of students and their details, then print the details.

•	Inputs: Number of students and their details (name, age, GPA).

•	Expected Output: Details of each student.

Medium


3. Dynamic 2D Array Allocation


•	Task: Write a program that dynamically allocates a 2D array of integers. Ask the user for the number of rows and columns, and then the elements of the 
array. Print the 2D array.

•	Inputs: Number of rows, number of columns, and the elements of the array.

•	Expected Output: The elements of the 2D array in matrix form.


4. Employee Records using Struct


•	Task: Create a struct called Employee with fields for ID, name, and salary. Dynamically allocate an array of Employee structs. Ask the user for the 
number of employees and their details. Write functions to:

o	Add a new employee

o	Delete an employee by ID

o	Print all employee details

•	Inputs: Number of employees, their details (ID, name, salary), operations to perform.

•	Expected Output: Updated employee records after each operation.

Hard


5. Dynamic Linked List of Books


•	Task: Create a linked list to store information about books. Each book has a title, author, and ISBN. Implement functions to:

o	Add a new book to the list

o	Remove a book by ISBN

o	Print all books in the list

•	Inputs: Operations to perform (add, remove), and book details (title, author, ISBN).

•	Expected Output: Updated list of books after each operation.


6. Dynamic Allocation and Struct for Library System


•	Task: Create a library system using dynamic allocation and structs. The library has an array of Book structs and an array of Member structs. Each Book 
has a title, author, and availability status. Each Member has a name, ID, and a list of borrowed books. Implement functions to:

o	Add new books to the library

o	Register new members

o	Borrow and return books

o	Print library and member details

•	Inputs: Number of books, number of members, their details, and operations to perform (add book, register member, borrow/return book).

•	Expected Output: Updated library and member details after each operation.




Linked List



Easy


1. Singly Linked List: Basic Operations


•	Task: Implement a singly linked list with basic operations: insertion at the beginning, insertion at the end, and printing the list.

•	Inputs: Values to insert.

•	Expected Output: The elements of the linked list after each insertion.


2. Singly Linked List: Deletion


•	Task: Extend the singly linked list implementation to include deletion of a node by value.

•	Inputs: Values to insert and values to delete.

•	Expected Output: The elements of the linked list after each deletion.

Medium


3. Singly Linked List: Search


•	Task: Add a function to search for a value in the singly linked list and return its position.

•	Inputs: Values to insert and values to search for.

•	Expected Output: The position of each searched value if found, or a message indicating that the value is not found.


4. Doubly Linked List: Basic Operations


•	Task: Implement a doubly linked list with basic operations: insertion at the beginning, insertion at the end, and printing the list both forwards and 
backwards.

•	Inputs: Values to insert.

•	Expected Output: The elements of the linked list after each insertion, printed both forwards and backwards.

Hard


5. Circular Linked List: Basic Operations


•	Task: Implement a circular linked list with basic operations: insertion at the beginning, insertion at the end, and printing the list.

•	Inputs: Values to insert.

•	Expected Output: The elements of the circular linked list after each insertion.


6. Linked List: Merge Two Sorted Lists

•	Task: Write a function to merge two sorted singly linked lists into a single sorted linked list.

•	Inputs: Two sorted linked lists.

•	Expected Output: The merged sorted linked list.



Advanced



7. Linked List: Detect and Remove Cycle

•	Task: Write a function to detect if a cycle exists in a singly linked list. If a cycle is detected, remove it.

•	Inputs: Values to insert (some of which form a cycle).

•	Expected Output: Indication of whether a cycle was detected and the elements of the linked list after removing the cycle.


8. Linked List: Reverse in Groups

•	Task: Write a function to reverse nodes in k-group in a singly linked list.

•	Inputs: Values to insert and the group size k.

•	Expected Output: The linked list after reversing in k-groups.




Stacks and Queues




1. Implementing a Stack


•	Task: Implement a stack using an array. Include basic operations such as push (inserting an element), pop (removing the top element), and peek (viewing the top element without removal).

•	Inputs: Elements to push and operations to perform (push, pop, peek).

•	Expected Output: Stack elements after each operation.


2. Valid Parentheses


•	Task: Write a function to determine if the parentheses in a given string are valid using a stack. Validity means every opening bracket has a corresponding closing bracket in the correct order.

•	Inputs: Strings with parentheses.

•	Expected Output: Whether the parentheses are valid or not.



3. Implementing a Queue


•	Task: Implement a queue using an array. Include basic operations such as enqueue (inserting an element), dequeue (removing the front element), and peek (viewing the front element without removal).

•	Inputs: Elements to enqueue and operations to perform (enqueue, dequeue, peek).
Expected Output: Queue elements after each operation.


4. Circular Queue


•	Task: Extend the queue implementation to create a circular queue. Implement basic operations (enqueue and dequeue) considering the circular nature of the queue.

•	Inputs: Elements to enqueue and operations to perform (enqueue, dequeue).

•	Expected Output: Queue elements after each operation.



Hard



5. Stack using Queues


•	Task: Implement a stack using two queues. Include operations such as push (inserting an element), pop (removing the top element), and top (viewing the top element without removal).

•	Inputs: Elements to push and operations to perform (push, pop, top).

•	Expected Output: Stack elements after each operation.


6. Design a Queue using Stacks


•	Task: Design a queue data structure using two stacks. Implement operations such as enqueue (inserting an element), dequeue (removing the front element), and front (viewing the front element without removal).

•	Inputs: Elements to enqueue and operations to perform (enqueue, dequeue, front).

•	Expected Output: Queue elements after each operation.



Advanced



7. Evaluate Reverse Polish Notation (RPN)


•	Task: Write a function to evaluate a given arithmetic expression in Reverse Polish Notation (RPN) using a stack.


•	Inputs: Arithmetic expressions in RPN format.

•	Expected Output: The result of the arithmetic expression.


8. Design a Priority Queue


•	Task: Design a priority queue using heaps (either min-heap or max-heap). Implement basic operations such as insert (enqueue with priority), remove (dequeue with highest priority), and peek (view element with highest priority).

•	Inputs: Elements with priorities and operations to perform (insert, remove, peek).

•	Expected Output: Queue elements after each operation.
